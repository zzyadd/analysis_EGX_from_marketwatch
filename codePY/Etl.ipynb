import the necessary libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
```

```python
url_page_1="https://www.marketwatch.com/tools/markets/stocks/country/egypt/1"
```

```python
response_page_1 = requests.get(url_page_1)
```

```python
soup_page_1 = BeautifulSoup(response_page_1.text, 'html.parser')
```

```python
# Find the table with class name "my-table"
tables= soup_page_1.find_all('table', class_='table-condensed')
```

```python
my_dict = {
    "Name": [],
    "Sector": [],
    "URL":[]
}
for table in tables:
    # Find all rows within the table body
    rows = table.find('tbody').find_all('tr')

    # Loop through the rows and extract data
    for row in rows:
        # Extract data from each column
        columns = row.find_all('td')
        if columns:
            name = columns[0].text.strip()
            sector = columns[2].text.strip()
            urll= columns[0].find("a").get("href")
            my_dict["Name"].append(name)
            my_dict["Sector"].append(sector)
            my_dict["URL"].append(f"https://www.marketwatch.com{urll}")
```

```python
url_page_2="https://www.marketwatch.com/tools/markets/stocks/country/egypt/2"
response_page_2 = requests.get(url_page_2)
soup_page_2 = BeautifulSoup(response_page_2.text, 'html.parser')
tables= soup_page_2.find_all('table', class_='table-condensed')
for table in tables:
    # Find all rows within the table body
    rows = table.find('tbody').find_all('tr')

    # Loop through the rows and extract data
    for row in rows:
        # Extract data from each column
        columns = row.find_all('td')
        if columns:
            name = columns[0].text.strip()
            sector = columns[2].text.strip()
            urll= columns[0].find("a").get("href")
            my_dict["Name"].append(name)
            my_dict["Sector"].append(sector)
            my_dict["URL"].append(f"https://www.marketwatch.com{urll}")
```

```python
df=pd.DataFrame(my_dict)
df
```

```python
URL2=[]
i=0
for li in df['URL']:
    url1=li
    response1 = requests.get(url1)
    soup1 = BeautifulSoup(response1.text, 'html.parser')
    try:
        # Attempt to find the first <a> element with the specified class and text
        link_element = soup1.find_all("a", class_="link" , text="Historical Quotes")
        
        if link_element:
            ok=link_element[0].get("href")
            print(i,ok)
        else:
            print("Element not found")
    except IndexError:
        print("IndexError: Element not found")
   
    URL2.append(ok)
    i=i+1
URL2
```
