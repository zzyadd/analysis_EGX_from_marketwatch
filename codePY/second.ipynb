download EGX data csv files

```python
for index, row in df.iterrows():
    url = row[3]
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup3 = BeautifulSoup(response.text, 'html.parser')
        csv_link=soup3.find_all("a",href=True,  class_="link link--csv m100")
        x=csv_link[0]['href'].split('startdate=')
        x[1] = 'startdate=09/23/2021' + x[1][10:]
        x[1]
        x=x[0]+x[1]
        csv_link[0]['href']=x
        csv_response=requests.get(csv_link[0]['href'])
        local_path = f"D:/EGX/{row[0].split('(')[1][:-1]}.csv"
        with open(local_path, 'wb') as file:
            file.write(csv_response.content)
        print(index," is Done")
    
```
